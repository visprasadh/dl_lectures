```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
torch.manual_seed(42)
```

```python
# Define the transformations
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.RandomCrop(32),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    transforms.ToTensor()
])
```

```python
# Load the CIFAR-10 training dataset with augmentations
trainset_original = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transforms.ToTensor())
trainset_augmented = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)

trainloader_original = torch.utils.data.DataLoader(trainset_original, batch_size=10, shuffle=False)
trainloader_augmented = torch.utils.data.DataLoader(trainset_augmented, batch_size=10, shuffle=False)
```

```python
# Helper function to show images
def imshow(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
for images_o, labels_o in trainloader_original:
    break

# Show images
imshow(torchvision.utils.make_grid(images_o))

# Print labels
print(labels_o)
```

```python
for images_a, labels_a in trainloader_augmented:
    break

# Show images
imshow(torchvision.utils.make_grid(images_a))

# Print labels
print(labels_a)
```
